```python
import pandas as pd
#Pandas is a Library used for data manipulation 
#Here Pandas Library is imported and refrenced as 'pd'
```

```python
dataset=pd.read_csv("Salary_Data.csv")
#here we use pd.read_csv to read the data from "Salary_Data.csv" file and this is saved under named 'dataset'
```

```python
dataset
#printing the values from the file to know if the file contents are read by library
```

```python
independent=dataset[["YearsExperience"]]
# As we have a preprocessed data we are assiging the input and output from the data
# we are assigning the input dataset to 'indepedent variable'
```

```python
independent # printing the assigned input dataset - Indepedent variable
```

```python
dependent=dataset[["Salary"]] ## we are assigning the output dataset to 'depedent variable'
```

```python
dependent # printing the assigned Output dataset - dependent variable
```

```python
from sklearn.model_selection import train_test_split
X_Train,X_Test,Y_Train,Y_Test=train_test_split(independent,dependent,test_size=0.30,random_state=0)


# After assiging the input and output from the preprocessed data we need to split the Training and Test data set for the model creation
# we are using sklearn.model_selection library and importing train_test_split -> which is used to split the training and Testing data from independent and dependent variables saved from previous steps
#train_test_split(<input>,<output>,<testdata percentage>,constantvalue)
# we are assinging train_test_split(independent,dependent,test_size=0.30,random_state=0) to X_Train,X_Test,Y_Train,Y_Test
# X_Train,X_Test are the training and Test data set for Xaxis
# Y_Train,Y_Test are the training and Test data set for Yaxis
```

```python
from sklearn.linear_model import LinearRegression
regressor=LinearRegression()
regressor.fit(X_Train,Y_Train)

#Now we are import LinearRegression function from sklearn.linear_model library
#This LinearRegression has the linear model way of predicting the data as per the linear model y=mx+c
#we are assigning LinearRegression() to regressor variable
#we are calling the Linearregression() function called 'fit' by regressor.fit(<input>,<output>)
#we are assinging the training data to regressor.fit(X_Train,Y_Train)
#we have loaded the procedure here , so we are subiting the training data as parameters to .fit(<Xaxis Trainingset>,<Yaxis Trainingset>
#at this step the learning process happens
#once executed by Run command <Linear Regression> function is called which is displayed below
```

```python
weight=regressor.coef_
weight

#Slope or Intercept value can be found from this step
#to the training data variable 'regressor' we are calling method regressor.coef_ which calculates the weight or slope
#y = mx+c <m - is the slope ir weight>
#calculation of slope determined here , my guess is this weight is nothing but every year the salry is getting risen by 9360.261 as per the weight predicted below
```

```python
bais=regressor.intercept_
bais

#Calculating the intercept value 'c' from y=mx+c
# (calculating the value of y when x = 0)
# so for this data set , it would be calculating the '0' years of experience here
##to the training data variable 'regressor' we are calling method regressor.intercept_ which calculates the bias
```

```python
y_pred=regressor.predict(X_Test)

#Now we are predicting the output using y_pred variable
#so here we are calling regressor.predict function and passing the paramter which is X_Test - xaxis test data now
#the predicted values for the test data inputs from X_test are saved in y_pred variable
#prediction is completed here
```

```python
from sklearn.metrics import r2_score
r_score=r2_score(Y_Test,y_pred)

#now the predicted values needs to be evaluated here
#we are using sklearn.metrics library and import r2_score function
#we are comparing the Y_Test data with y_pred and calculating the score , if the score is

# 1	    Perfect prediction (model explains 100% of the variance)
# 0	    Model does no better than simply predicting the mean
# < 0	Model performs worse than predicting the mean 
```

```python
r_score
# Our model prediction is printed as 97# which is almost equal to perfect prediction
# Once the evaluation is complete we need to Save the model
```

```python
import pickle
filename="finalized_model_linear.sav"

#importing pickle library to save the predicted model in a filename called "finalized_model_linear.sav"
#pickle filename extension is .sav <filename.sav>
```

```python
pickle.dump(regressor,open(filename,'wb'))

#calling pickle.dump (<trained model> which is regressor, adding this trained to filename in 'wb' mode - writebinary mode)
#Once the model is saved a file is added in directory under the name "finalized_model_linear.sav"
```

```python
loaded_model=pickle.load(open("finalized_model_linear.sav",'rb'))
result=loaded_model.predict([[21]])

#This step is to load the predicted modal and add input to predicted modal and check the results
#function called pickle.load
#pickle.load(open("finalized_model_linear.sav",'rb')) -> rb read binary mode
# saving the pickle.load under variable loaded_model
# in order to add the inputs add loaded_model.predict([[21]]) where '21' is the input number
#save the loaded_model.predict to result variable
```
